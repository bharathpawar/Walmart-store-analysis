# **Visulization using  LIBRARY Seaborn ****. **


```
import seaborn as sns
```

```
sns.distplot(df.Item_Weight,bins=10)
```

# HERE IN THE ,
# DISTRIBUTION PLOT FOR THE VARIABLE  ITEM_WEIGHT  WE CAN UNDERSTAND HOW THE VALUES ARE DISTRIBUTED and we can know the  values of item weight ...HOW THE VARIABLE IS PLOTTED ETC..

```
sns.barplot(x='Item_Weight',y='Item_Fat_Content',palette='Set1',data=df)
```

#almost every bar is same as the other FOR ITEM_FAT_CONTENT WITH RESPECT TO ITEM_WEIGHT  so , here we cant clearly interpret so below is the graph using matplotlib for better understanding>.....

```
import matplotlib.pyplot as plt # To interpret the above chart we need the help of Library (matplotlib) ....
```

```
plt.bar(df.Item_Weight,df.Item_Fat_Content,color='red',hatch='.',edgecolor='green')
```

#HERE IN THE GRAPH,
# all most all the weights for the different types of fat are probably same with respect to weight so we can see the deviation of  every item_fat_content

```
plt.bar(df.Item_Weight,df.Item_Type,color='green',hatch='.',edgecolor='green') 
```

# HERE IN THE GRAPH,
# we can see the data for item _weight and item_type and the values ,here we cant clearly interpret so To understand easily above graph is plotted below using plotly . express ....( )

```
import plotly.express as px # To interpret the above chart we need the help of library (plotly.express)...
```

```

px.bar(df.Item_Weight,df.Item_Type,barmode='group',color_discrete_sequence=["white"],template="plotly_dark")
```

# HERE IN THE GRAPH
# here can we can easily understand the graph based on the item weight and item type.. and we can see the data for the given  variables

```
px.scatter(df,x='Item_Weight',y='Item_Type',color='Outlet_Location_Type',template="plotly_dark",animation_frame="Outlet_Establishment_Year")
```

# HERE IN THE GRAPH WE CAN GET THE DATA OF  OUTLET_LOCATION  FOR  DIFFERENT ITEM_TYPES WITH RESPECT TO ITEM _WEIGHT and also WE CAN SEE  THE CHANGES OF HOW VALUES ARE CHANGING FOR THE VARIABLE YEAR 

```
px.bar(df,x='Item_Type',y='Item_MRP',color='Outlet_Size',barmode='group',color_discrete_sequence=["blue"],template="plotly_dark",animation_frame="Outlet_Establishment_Year")
```

# Here,we can know  mrp prize for item_type  depending upon the outlet_size so we can take some better business decisions for manufacturing with respect to prize  WE CAN ALSO SEEE THE VALUES ARE CHANGING ACCORDING TO YEAR

```
px.pie(df,values='Item_MRP',names='Item_Type',color='Outlet_Location_Type',template="plotly_dark")
```

# HERE ,WE CAN SEE THAT ITEM MRP FOR THE DIFFERENT TYPES WITH RESPECT TO OUTLET_LOCATION_TYPE .. SO ,WE CAN UNDERSTAND THAT RESPECTIVE ITEM IS LOCATED IN RESPECTIVE LOCATION (TYPE) ! 
# ADAVANTAGE IS WE CAN KNOW THE ITEM_MRP ALSO....

```
px.pie(df,values='Item_Outlet_Sales',names='Item_Fat_Content',color='Outlet_Location_Type',template="plotly_dark")
```

# HERE WE CAN UNDERSTAND THESE LOWFAT,REGULAR,LF,REG,LOWFAT ARE RELATED TO WHICH OUTLET_LOCATION AND WITH RESPECT TO SUM OF THEIR ITEM_OUTLET_SALES ,,SO HERE WE CAN TAKE A  BUSINESS DECISION BASED ON THE SUM OF SALES FOR  DIFFERENT TYPES OF ITEM_FAT_CONTENT
